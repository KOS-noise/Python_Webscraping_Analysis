#### 3-1. 네이버 책 검색 API 호출하기


```python
from dotenv import load_dotenv
import os
import requests
```

```python
# env 파일을 읽어오자
load_dotenv()

CLIENT_ID = os.getenv("CLIENT_ID")
print(CLIENT_ID[:4]) # 글자 인덱스 0부터 3까지

CLIENT_SECRET = os.getenv("CLIENT_SECRET")
print(CLIENT_SECRET[:4])

# env 파일에 있는 클라이언트 아이디,시크릿을 변수로 불러왔으므로 앞으로의 모든 코드에서 복붙할 필요가 없음.
```

```python
def search_books(query):  # 검색어(query)는  함수의 인자로 받아서 동적으로 처리 되어야 합니다. 

    headers = {
    'X-Naver-Client-Id': CLIENT_ID,
    'X-Naver-Client-Secret': CLIENT_SECRET,
    }   
    
    payload = {
        'query': query,
        'display': 100,
        'sort': 'sim'
    }
    
    url = 'https://openapi.naver.com/v1/search/book.json'
    
    # requests get(url, params, headers) 요청 
    res = requests.get(url, params=payload, headers=headers)
    
    
    # json() 함수로 응답 결과 가져오기 [{}, {}, {}]
    if res.status_code == 200:
        items_data = res.json()['items'] 
        return items_data # 추가 (셀분리)
    else:
        print("Error Code:", res.status_code)
        return []
```

```python
search_books('파이썬')
```

```python
# 질문 1 :  검색어로  찾은  책 목록을 json 파일로 저장하기
import json

# 앞에서 만든 함수로 데이터 먼저 가져오기
items_data = search_books('파이썬')

# 현재 위치가 data 랑 다름 
with open('../data/books.json', 'w', encoding='utf-8') as file:
    json.dump(items_data, file)
```

```python
# 2. books.json 파일을 Pandas DataFrame로 저장하기

import pandas as pd

book_df = pd.read_json('../data/books.json')
```

```python
# 3. 질문 :  검색어로  찾은  책 목록 출력하기
book_df.head(3)
```

```python
# 4. 질문 :  검색어로  찾은  책 목록 중에서 가격이 2만원 이상인 책만 출력하기
book_df[book_df['discount'] >= 20000][['title', 'author', 'discount', 'publisher', 'pubdate']]\
    .sort_values(by='discount', ascending=False)\
    .reset_index(drop=True)\
    .head(10)
```

```python
# 5. 질문 :  검색어로  찾은  책 목록 중에서 출판사가 “인피니티북스”인 책만 출력하기
book_df[book_df['publisher'] == '인피니티북스'].drop(columns=['image', 'description']).reset_index(drop=True)
```
