#### 1-1 Daum 뉴스기사 제목 스크래핑하기

````python
import requests
import bs4
````

````python
from bs4 import BeautifulSoup
````

````python
# url 변수 선언
# url = url = 'https://news.daum.net/economy'

req_param = { 'sid' : 'economy'}
url = 'https://news.daum.net/{sid}'.format(**req_param)
print(url)
````

````python
req_header = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36' 
}
res = requests.get(url, headers = req_header)
print(type(res))
````

````python
res.encoding = 'utf-8' # 한국어 인코딩
print(res.status_code)
````

```
<ul class="list_newsheadline2">
   <li> 
     <a href="링크">
     <div class="cont_thum">
        <strong class="tit_txt">뉴스제목</strong>
        
```

````python
if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')
    li_tags = soup.select('ul.list_newsheadline2 li')

    print(type(li_tags), len(li_tags))

    for li_tag in li_tags:
        a_tag = li_tag.find('a') # 먼저 a 태그 찾기
        if a_tag:
            link = a_tag['href'] # link = href 속성
            strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt') # div > strong

            if strong_tag:
                title = strong_tag.text.strip()  # 뉴스 제목만 추출

                print(link)
                print(title)
else:
    print(f"Error Code: {res.status_code}")
````

##### 질문2:  여러개의 section 중 하나를 선택해서 url에서 뉴스기사의 링크와 제목을 출력하는 코드를 함수로 작성하기


````python
# url은 위에서 선언함 
section_dict = {'기후/환경': 'climate','사회': 'society','경제': 'economy',
                '정치':'politics','국제':'world','문화': 'culture', '생활' : 'life',
                'IT/과학' : 'tech', '인물' : 'people'}
````

````python
# 함수 선언
def print_news(section_name):
    sid = section_dict.get(section_name, 'IT/과학')

    url = f'https://news.daum.net/{sid}'

    req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.3'
    }
    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8' # res 선언 밑에다가 
    # requests가 가끔 한글 사이트의 인코딩을 자동으로 못 찾을 때가 있어서,

    if res.ok:
        soup = BeautifulSoup(res.text,'html.parser')
        print(f"======> {url} {section_name} 뉴스 <======")

        li_tags = soup.select('ul.list_newsheadline2 li')

        for li_tag in li_tags:
            a_tag = li_tag.find('a')
            if a_tag:
                link = a_tag['href']
                strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')

                if strong_tag:
                    title = strong_tag.text.strip()
                    print(link)
                    print(title)
    else:
            print(f"Error Code: {res.status_code}")
                

    
````

````python
# 함수 호출 
print_news('문화')
````
