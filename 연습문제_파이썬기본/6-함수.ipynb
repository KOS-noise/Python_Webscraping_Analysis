```python
# 6-1
# 설명: 두 수를 받아서 사칙연산을 수행하는 함수들을 작성하는 프로그램을 만드세요

def add(a,b):
    return a + b

def sub(a, b):
    return a-b

def mul(a, b):
    return a * b

def divide(a, b):
    return a / b

print(f"10 + 5 = {add(10, 5)}")
print(f"10 - 5 = {sub(10, 5)}")
print(f"10 * 5 = {mul(10, 5)}")
print(f"10 / 5 = {divide(10, 5):.1f}")
```

```python
# 6-2
# 설명: 팩토리얼을 계산하는 재귀함수와 반복문을 사용한 함수를 각각 작성하세요

def factorial_recursive(n):
    if n == 1:
        return 1
    return n * factorial_recursive(n -1) # 자기 자신을 한번 더 불러온다 

def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(f"5! (재귀) = {factorial_recursive(5)}")
print(f"5! (반복) = {factorial_iterative(5)}")
print(f"7! (재귀) = {factorial_recursive(7)}")
print(f"7! (반복) = {factorial_iterative(7)}")
```

```python
# 6-3
# 설명: 기본값 매개변수를 사용하여 인사말을 생성하는 함수를 작성하세요

def greet(name, greeting = "안녕하세요", suffix = ""):  # 가변 매개변수는 제일 마지막에 온다 ~
    if "Hello" in greeting:
        print(f"{greeting}, {name}!")
    else:
        # 추가 멘드 (suffix)가 있는 경우
        if suffix:
            print(f"{greeting}, {name}님! {suffix}")
        else:
            print(f"{greeting}, {name}님!")

greet("김철수")  # 기본값 사용
greet("John", "Hello")  # 인사말 변경
greet("이영희", suffix="좋은 하루 되세요!")  # 추가 멘트 사용
```

```python
# 6-4
# 설명: 리스트의 통계 정보(평균, 최댓값, 최솟값, 표준편차)를 반환하는 함수를 작성하세요

import math

def calculate(numbers):
    mean = sum(numbers) / len(numbers)
    max_val = max(numbers) 
    min_val = min(numbers)

    var = sum([(x - mean) ** 2 for x in numbers]) / len(numbers)
    std_dev = math.sqrt(var)

    return mean, max_val, min_val, std_dev

data = [10, 20, 30, 40, 50]

mean, max_val, min_val, std_dev = calculate(data)

print(f"숫자들: {data}")
print(f"평균: {mean:.1f}")
print(f"최댓값: {max_val}")
print(f"최솟값: {min_val}")
print(f"표준편차: {std_dev:.2f}")
```

```python
# 6-5
# 설명: 람다 함수를 사용하여 리스트를 다양한 방식으로 정렬하는 프로그램을 작성하세요

students = [('김철수', 85),
            ('이영희', 92),
            ('박민수', 78),
            ('최수진', 95)]

# 튜플의 0번째 요소 기준
sorted_by_name = sorted(students, key = lambda x: x[0])
# 튜플의 1번째 요소 기준
sorted_by_score = sorted(students, key = lambda x: x[1])

sorted_by_score_desc = sorted(students, key = lambda x: x[1], reverse=True)

print(f"학생 정보: {students}")
print(f"이름순 정렬: {sorted_by_name}")
print(f"점수순 정렬: {sorted_by_score}")
print(f"점수 내림차순: {sorted_by_score_desc}")
```
