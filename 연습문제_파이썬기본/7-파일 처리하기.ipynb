```python
# 7-1
# 설명: 텍스트 파일에 여러 줄의 문자열을 저장하고 읽어오는 프로그램을 작성하세요

filename = "example.txt"
lines = [
    "안녕하세요\n",
    "파이썬 파일 처리를 연습하고 있습니다\n",
    "오늘은 좋은 날씨입니다\n"
]

with open(filename, "w", encoding="utf-8") as f:
    for line in lines:  
        f.write(line)   


print("파일에 저장할 내용:")
for line in lines:
    print(line.strip())

print("\n")
print("파일에서 읽어온 내용:")
with open(filename, "r", encoding="utf-8") as f:
    read_lines = f.readlines()
    for line in read_lines:
        print(line.strip())
```

```python
# 7-2
# 설명: CSV 형태의 학생 성적 데이터를 파일에 저장하고 읽어서 평균을 계산하는 프로그램을 작성하세요.

import csv

filename = "grades.csv"
header = ["이름", "점수"]
data = [
    ["김철수", 85],
    ["이영희", 92],
    ["박민수", 78],
    ["최수진", 95]
]

with open(filename, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data)

print(f"학생 성적이 {filename}에 저장되었습니다.")

print("성적 분석 결과:")
total_score = 0
count = 0

with open(filename, "r", encoding = "utf-8") as f:
    reader = csv.reader(f)
    next(reader)

    for row in reader:
        name = row[0]
        score = int(row[1])
        print(f"{name}: {score}점")
        total_score += score
        count += 1


average = total_score / count
print(f"전체 평균: {average:.1f}점")
```

```python
# 7-3
# 설명: 텍스트 파일의 단어 빈도를 계산하는 프로그램을 작성하세요.


from collections import Counter

filename = "words.txt"
content = "파이썬 프로그래밍 언어 파이썬 배우기 쉬운 언어 파이썬 강력한 프로그래밍"

with open(filename, "w", encoding="utf-8") as f:
    f.write(content)

with open(filename, "r", encoding="utf-8") as f:
    text = f.read()
    words = text.split()
    word_counts = Counter(words)

print("단어 빈도 분석 결과:")

for word, count in word_counts.most_common():
    print(f"{word}: {count}번")
```

```python
# 7-4
# 설명: JSON 형태의 데이터를 파일에 저장하고 읽어오는 프로그램을 작성하세요
import json

filename = "data.json"
data = {
    "이름": "김철수",
    "나이": 25,
    "직업": "개발자",
    "취미": ["독서", "영화감상", "코딩"],
    "주소": "서울시 강남구"
}


with open(filename, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print(f"데이터가 {filename}에 저장되었습니다.")

print("JSON에서 읽어온 데이터:")
with open(filename, "r", encoding="utf-8") as f:
    loaded_data = json.load(f)
 
    for key, value in loaded_data.items():
        
        if isinstance(value, list):
            print(f"{key}: {value}") 
        else:
            print(f"{key}: {value}")
```

```python
# 7-5
# 설명: 로그 파일을 생성하고 특정 레벨의 로그만 필터링하여 출력하는 프로그램을 작성하세요.

filename = "app.log"
logs = [
    "2025-07-20 09:15:00 - WARNING - 메모리 사용량이 높습니다\n",
    "2025-07-20 10:00:00 - INFO - 서버 시작\n",
    "2025-07-20 10:30:00 - ERROR - 데이터베이스 연결 실패\n",
    "2025-07-20 11:45:00 - ERROR - 파일을 찾을 수 없음\n",
    "2025-07-20 12:00:00 - WARNING - 디스크 공간 부족\n",
    "2025-07-20 12:30:00 - INFO - 사용자 로그인\n"
]

# 1. 로그 파일 생성
with open(filename, "w", encoding="utf-8") as f:
    f.writelines(logs)

print("로그 파일이 생성되었습니다.")

# 2. 로그 필터링 및 출력
with open(filename, "r", encoding="utf-8") as f:
    lines = f.readlines()
    
    # ERROR 레벨 출력
    print("ERROR 레벨 로그들:")
    for line in lines:
        if "ERROR" in line:
            print(line.strip())

    # WARNING 레벨 출력
    print("WARNING 레벨 로그들:")
    for line in lines:
        if "WARNING" in line:
            print(line.strip())
```
