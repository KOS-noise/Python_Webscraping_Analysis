```python
# 4-1
score = int(input("점수를 입력하세요:"))

if score >= 90:
    grade='A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f"{score}점의 학점은 {grade}입니다.")
```

```python
# 4-2
three = []
sum=0
for i in range(1,101):
    if i % 3 ==0:
        three.append(i)
        sum += i

print(f"1부터 100까지 3의 배수: {three}")



print(f"3의 배수의 합 : {sum}")
print(f"3의 배수의 개수 : {len(three)}")
```

```python
# 4-3
d = int(input("몇 단을 출력할까요? "))
print(f"{d}단 구구단")

for i in range(1, 10):
    print(f"{d} x {i} = {d * i}")
```

```python
# 4-4
total = 0

while True:
    num=int(input("숫자를 입력하세요 (0을 입력하면 종료): "))

    if num == 0:
        break

    total += num

print(f"입력된 숫자들의 합: {total}")
```

```python
# 4-5
num = int(input("숫자를 입력하세요: "))
is_prime = True

# 1 이하의 숫자는 소수가 아님
if num < 2:
    is_prime = False
else:
    #2부터 자기 자신 전까지 나누어 떨어지는지 확인(제곱근까지만 확인해도됨)
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{num}은 소수입니다.")
else:
    print(f"{num}은 소수가 아닙니다.")
```
