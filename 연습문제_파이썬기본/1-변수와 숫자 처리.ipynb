```python
#  1-1

num1 = int(input("첫 번째 숫자를 입력하세요: "))
num2 = int(input("두 번째 숫자를 입력하세요: "))

print(f"첫 번째 숫자를 입력하세요: {num1}")
print(f"두 번째 숫자를 입력하세요: {num2}")

print(f"{num1} + {num2} = {num1 + num2}")
print(f"{num1} - {num2} = {num1 - num2}")
print(f"{num1} * {num2} = {num1 * num2}")
print(f"{num1} / {num2} = {num1 / num2:.2f}")
```

```python
# 1-2
PI = 3.14159
radius = int(input("원의 반지름을 입력하세요: "))
print(f"원의 반지름을 입력하세요: {radius}")
print(f"반지름이 5인 원의 넓이: {radius * radius * PI:.2f}")
print(f"반지름이 5인 원의 둘레: {radius * 2 * PI:2f}")
```

```python
# 1-3
celsius = float(input("섭씨 온도를 입력하세요: "))
fahrenheit = celsius * 9 / 5 + 32
print(f"섭씨 {celsius:.0f}도는 화씨 {fahrenheit:.1f}도입니다.")
```

```python
# 1-4
width = int(input("가로 길이를 입력하세요: "))
height = int(input("세로 길이를 입력하세요: "))

area = width * height
dec = (width + height) * 2

print(f"직사각형의 넓이: {area}")
print(f"직사각형의 둘레: {dec}")
```

```python
price = int(input("상품 가격을 입력하세요: "))
sale = int(input("할인율을 입력하세요(%): "))

print(f"원래 가격: {price}")
print(f"할인율: {sale}%")
print(f"할인 금액: {int(price * sale /100)}")
print(f"최종 가격: {int(price - (price*sale/100))}")
```
