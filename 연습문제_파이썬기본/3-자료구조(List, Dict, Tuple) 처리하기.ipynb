```python
# 3-1
numbers = [10, 20, 30, 40, 50]

total = sum(numbers)
average = total / len(numbers)

print(f"숫자들: {numbers}")
print(f"합계: {int(total)}")
print(f"평균: {average}")
```

```python
# 3-2
fruits = ['사과', '바나나', '오렌지', '포도', '딸기']

print(f"과일 목록: {fruits}")
search_fruit = input("찾을 과일을 입력하세요: ")

if search_fruit in fruits:
    print(f"'{search_fruit}'가 목록에 있습니다!")
```

```python
# 3-3
grades = {'김철수': 85, '이영희': 92, '박민수': 78, '최수진': 95}

print("학생 성적: ")
total_score = 0

for name, score in grades.items():
    print(f"{name}: {score}점")
    total_score += score

average = total_score / len(grades)
print(f"평균 점수: {average}점")
```

```python
# 3-4
import math

point1 = (0,0)
point2 = (3,4)

distance = math.sqrt((point2[0]-point1[0])**2 + (point2[1] - point1[1])**2)

print(f"점 1: {point1}")
print(f"점 2: {point2}")
print(f"두 점 사이의 거리: {distance}")
```

```python
# 3-5
cart = {
    '사과': {'price': 1000, 'n': 2},
    '바나나': {'price': 800, 'n': 3},
    '오렌지': {'price': 1500, 'n': 1}
}

print("쇼핑 카트:")
total_price = 0

for item, info in cart.items():
    subtotal = info['price'] * info['n']
    total_price += subtotal
    print(f"{item}: {info['n']}개 (개당 {info['price']}원) = {subtotal}원")

print(f"총 가격: {total_price}원")
```

```python
# 3-6
numbers = [15, 3, 27, 8, 19, 12, 31]

# 정렬을 통해 두 번째로 큰 값 찾기
sorted_numbers = sorted(numbers, reverse=True) # 내림차순

print(f"숫자 목록: {numbers}")
print(f"최댓값: {max(numbers)}")
print(f"최솟값: {min(numbers)}")
print(f"두 번째로 큰 값: {sorted_numbers[1]}")
```

```python
# 3-7
words = ['cat', 'elephant', 'dog', 'butterfly', 'ant']

longest = max(words, key=len)
shortest = min(words, key=len)

print(f"단어 목록: {words}")
print(f"가장 긴 단어: {longest} ({len(longest)}글자)")
print(f"가장 짧은 단어: {shortest} ({len(shortest)}글자)")
```

```python
# 3-8
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
unique = sorted(list(set(numbers)))

print(f"원본 리스트: {numbers}")
print(f"중복 제거 후: {unique}")
```

```python
# 3-9
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
print(f"리스트1: {list1}")
print(f"리스트2: {list2}")

merged_list = list(set(list1 + list2))
print(f"병합된 리스트: {merged_list}")

common = []

for item in list1:
    if item in list2:
        common.append(item)

print(f"공통 요소: {common}")
```

```python
# 3-10
students = [
    {'name' : '김철수', 'age' : 20, 'major' : '컴퓨터공학과'},
    {'name' : '박민수', 'age' : 21, 'major' : '경영학과'},
    {'name' : '이영희', 'age' : 22, 'major' : '영어영문학과'},
    {'name' : '최수진', 'age' : 23, 'major' : '수학과'}
]

sorted_students = sorted(students, key = lambda x: x['age'])

print("나이 순으로 정렬된 학생 목록")
for student in sorted_students:
    print(f"{student['name']} ({student['age']}세) - {student['major']}")
```
