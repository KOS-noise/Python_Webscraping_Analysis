```python
# 8-1
# 설명: 수학 연산을 위한 모듈을 만들고 이를 사용하는 메인 프로그램을 작성하세요.
# 파일명 : math_operations.py

PI = 3.14159

def circle_area(radius):
    return PI * radius * radius  # 원의 넓이 계산

def rectangle_area(width, height):
    return width * height   # 직사각형 넓이 계산 

def factorial(n):
    result = 1

    for i in range(1, n+1):
        result = result * i
    return result

def cho(a, b):
    # b가 0이 될 때까지 반복
    while b > 0:
        a, b = b, a%b  # a에는 b값을, b에는 a를 b로 나눈 나머지를 대입
    return a
```

```python
# 8-2
# 설명: datetime과 random 모듈을 사용하여 임의의 날짜와 숫자를 생성하는 프로그램을 작성하세요.

import datetime
import random

# datetime 모듈 사용
now = datetime.datetime.now()
print(f"현재 날짜와 시간: {now.strftime('%Y-%m-%d %H:%M:%S')}")

# 포맷된 날짜(ex : 2025년 07월 20일 일요일)
# 요일 처리를 위해 리스트를 활용한다. 월요일은 0, 일요일은 6
days = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
target_date = datetime.datetime(2025, 7, 20, 14, 30, 25)
day_str = days[target_date.weekday()]
print(f"포맷된 날짜: {target_date.strftime('%Y년 %m월 %d일')} {day_str}")

# random 모듈 사용
rand_int = random.randint(1, 10)
print(f"임의의 숫자 : {rand_int}")

# 임의의 실수
rand_float = random.random() * 5  # 0.0 ~ 5.0
print(f"임의의 실수: {rand_float:.2f}")

# 임의의 리스트 요소 선택 및 섞기
fruits = ['사과', '바나나', '오렌지', '포도', '딸기']
chosen_fruit = random.choice(fruits)
print(f"임의의 리스트 요소: {chosen_fruit}")

random.shuffle(fruits)
print(f"섞인 리스트: {fruits}")

```
